```python
import cv2
import unicodedata
import glob
```

```python
import torch

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5x')  # or yolov5m, yolov5l, yolov5x, custom
```

```python
import glob
gb = glob.glob('crop할 한국 이미지(차량의) 브랜드 경로1/*/*/*/*/*/*/*')

num = 0
for img in gb:
    
    # inference
    results = model(img)

    # Results
    crops = results.crop()

    max_area = 0
    max_img = None

    for crop in crops:
        current_area = crop['im'].shape[0] * crop['im'].shape[1]
        if max_area < current_area:
            max_img = crop['im']
            max_area = current_area
    num+=1
    file_name = img[img.rfind('/')+1:-4]
    file_name = unicodedata.normalize('NFKC', file_name)
    

    cv2.imwrite('저장경로'+file_name+'.jpg', max_img)
    print(file_name)
    print(num)    
```
